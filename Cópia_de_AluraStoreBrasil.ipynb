### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

```python
for i, loja in enumerate([loja1, loja2, loja3, loja4], start=1):
    print(f"\n Loja {i}")
    print(f"Linhas: {loja.shape[0]} | Colunas: {loja.shape[1]}")
    display(loja.head(2))
```

```python
loja1["Loja"] = "Loja 1"
loja2["Loja"] = "Loja 2"
loja3["Loja"] = "Loja 3"
loja4["Loja"] = "Loja 4"
```

```python
dados = pd.concat([loja1, loja2, loja3, loja4])
dados.head()
```

```python
print("Lojas disponíveis:", dados["Loja"].unique())

```

```python
for loja in dados["Loja"].unique():
    print(f"\n Amostra de registros da {loja}")
    display(dados.query(f"Loja == '{loja}'").head(3))
```

#1. Análise do faturamento



```python
faturamento_total = dados["Preço"].sum()
print(f" Faturamento total das 4 lojas: R$ {faturamento_total:,.2f}")
```

```python
faturamento_por_loja = dados.groupby("Loja")["Preço"].sum().sort_values(ascending=False)
print(" Faturamento por loja:")
display(faturamento_por_loja)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(faturamento_por_loja.index, faturamento_por_loja.values)
plt.title("Faturamento por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# 2. Vendas por Categoria

```python
faturamento_categoria = (
    dados.groupby("Categoria do Produto")["Preço"].sum().sort_values(ascending=False))

print(" Faturamento por categoria:")
display(faturamento_categoria)
```

```python
plt.figure(figsize=(10,6))
plt.barh(faturamento_categoria.index, faturamento_categoria.values, color='skyblue', alpha=0.7)
plt.title("Faturamento por Categoria de Produto")
plt.xlabel("Faturamento (R$)")
plt.ylabel("Categoria do Produto")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
categorias_principais = ["eletronicos", "eletrodomesticos", "moveis"]
dados_filtrados = dados[dados["Categoria do Produto"].isin(categorias_principais)]
```

```python
vendas_por_categoria_loja = (dados_filtrados.groupby(["Loja", "Categoria do Produto"]).size().reset_index(name="Quantidade"))

vendas_por_categoria_loja = vendas_por_categoria_loja.sort_values(by="Quantidade", ascending=False)
display(vendas_por_categoria_loja)
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.barplot(data=vendas_por_categoria_loja, x="Categoria do Produto", y="Quantidade", hue="Loja", palette=["#4B8BBE", "#FFC300", "#FF5733", "#82E0AA"])

plt.title("Vendas por Categoria (Eletrônicos, Eletrodomésticos e Móveis) por Loja", fontsize=14, fontweight="bold")
plt.xlabel("Categoria do Produto")
plt.ylabel("Quantidade de Vendas")
plt.legend(title="Loja")
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao = (dados.groupby("Loja")["Avaliação da compra"].mean().round(2).sort_values(ascending=False))
print(media_avaliacao)
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_produtos = (dados["Produto"].value_counts().reset_index().rename(columns={"index": "Produto", "Produto": "Quantidade de Vendas"}))

print("Produtos mais vendidos:")
print(vendas_produtos.head(5))

print("\n Produtos menos vendidos:")
print(vendas_produtos.tail(5))
```

# 5. Frete Médio por Loja

```python
frete_medio = (dados.groupby("Loja")["Frete"].mean().round(2).sort_values(ascending=False))

print("Frete médio por loja:")
print(frete_medio)
```

```python
cores = ["#4B8BBE", "#FFC300", "#FF5733", "#82E0AA"]

plt.figure(figsize=(5,5))
plt.pie(frete_medio, labels=frete_medio.index,autopct='%1.1f%%', startangle=90, colors=cores,wedgeprops={"edgecolor": "white", "linewidth": 2})

plt.title("Distribuição do Frete Médio por Loja", fontsize=14, fontweight="bold")
plt.show()
```

**3. Interpretação e implicações**

•	Por que a Loja 4 é candidata à venda?

- Menor faturamento é o critério objetivo mais forte para redução de portfólio.
- Frete médio menor pode indicar público/área com menor disposição para pagar ou estratégia de preços agressiva que reduz margem.
- Ficar em 2º lugar em eletrônicos e móveis mostra que há potencial, porém se o objetivo é simplificar e vender ativos de baixo desempenho, a Loja 4 é a escolha natural.

•	**Contrapontos / motivos para considerar NÃO vender imediatamente**

- Se a Loja 4 tem margens brutas maiores (mesmo com faturamento menor), ela pode ser mais lucrativa por venda.
- Se está em crescimento acelerado (tendência positiva), vender agora pode significar perda de resultado positivo.
- Baixo frete pode ser porque a loja opera em região com custos logísticos melhores — um comprador pode ver isso como vantagem e pagar bem.
- Aspectos não analisados: custos fixos (aluguéis, salários), dívidas, estoque encalhado, contratos locais, capacidade de escala.


**4. Recomendação prática**

Recomendação provisória: vender a Loja 4, mas somente após as validações do checklist financeiro-operacional abaixo. A decisão final deve incorporar margem, tendência de receita e riscos contratuais.


**5. Checklist obrigatório antes da venda (fatores que precisam ser verificados)**

1.	Margem bruta e líquida por loja (não apenas faturamento).
2.	Tendência de receita (últimos 6–12 meses) — crescimento, estagnação ou queda?
3.	Custo fixo mensal (aluguel, salários, utilidades) e EBITDA por loja.
4.	Inventário — valor de estoque (rotatividade / idade / obsolescência).
5.	Dívidas e passivos vinculados à loja (empréstimos, garantias).
6.	Contratos e obrigações (fornecedores, locação, funcionários).
7.	Retenção de clientes — se a Loja 4 tem clientes fiéis, isso aumenta valor.
8.	Risco operacional/regulatório (licenças, taxas locais).
9.	Motivo estratégico — vender para reduzir complexidade operacional ou para levantar caixa?
10.	Multiplicadores de mercado — avaliar preço por múltiplo de EBITDA se possível.

Só com faturamento e avaliações você tem boas pistas, mas a avaliação de venda justa precisa desses itens financeiros.


**6. Alternativa de decisão (se quiser priorizar outro critério)**

- Se prioridade = maximizar caixa imediato: vender a loja com menor faturamento e menor crescimento (Loja 4).

•	Se prioridade = reduzir risco de reputação: considerar vender a Loja 1 (avaliação média abaixo de 4), se reputação / avaliações forem críticas.

•	Se prioridade = manter mix de produtos fortes: manter lojas que são líderes nos produtos mais vendidos (identificar quais lojas dominam os 5 produtos mais vendidos).
